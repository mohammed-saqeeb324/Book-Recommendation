```python
def recommend_books(User_ID,user_sim_df,sample,num_recommendations=5):
    # Find the most similar user to the target user
    similar_users = user_sim_df.loc[User_ID].sort_values(ascending=False).index[0:]  # Exclude the user itself
        # books rated by the target user
    books_read_by_user = sample[sample['User_ID'] == User_ID]['Book_Title'].tolist()
    recommendations=[]
    for similar_user in similar_users:
        books_read_by_similar_user= sample[sample['User_ID'] == (similar_user)]['Book_Title'].tolist()
        for Book_Title in books_read_by_similar_user:
            if Book_Title not in books_read_by_user and Book_Title not in recommendations:
                recommendations.append(Book_Title)
                if len(recommendations) == num_recommendations:
                    return recommendations
    return recommendations                
    
```

```python
user_id_to_recommend=int(input("enter user id"))
recommendations=recommend_books(user_id_to_recommend,user_sim_df,sample,5)
print(f"Recommended books for user {user_id_to_recommend}: {recommendations}")
```
